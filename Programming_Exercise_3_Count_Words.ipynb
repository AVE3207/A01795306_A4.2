```python
program_code = """
\"\"\"
Módulo para contar la frecuencia de palabras en archivos de texto y escribir
los resultados en un archivo de salida.
\"\"\"

import time

def read_file(filename):
    \"\"\"Lee el contenido del archivo y maneja errores de lectura.\"\"\"
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            content = file.read()
        return content
    except IOError as e:
        print(f"Error leyendo el archivo {filename}: {str(e)}")
        return ""

def count_words(text):
    \"\"\"Cuenta la frecuencia de cada palabra única en el texto proporcionado.\"\"\"
    word_frequency = {}
    words = text.split()

    for word in words:
        word = ''.join([char for char in word if char.isalnum() or char == '-']).lower()
        if word:
            word_frequency[word] = word_frequency.get(word, 0) + 1

    return word_frequency

def write_results(filename, word_count, elapsed_time):
    \"\"\"Escribe los resultados del conteo de palabras en el archivo especificado.\"\"\"
    grand_total = sum(word_count.values())

    with open(filename, 'w', encoding='utf-8') as file:
        file.write("Word Count Results:\\n")
        for word, count in sorted(word_count.items(), key=lambda x: (-x[1], x[0])):
            file.write(f"{word}: {count}\\n")
        file.write(f"\\nGrand Total: {grand_total}\\n")
        file.write(f"Time Elapsed: {elapsed_time:.4f} seconds\\n")

def process_file(input_filename, output_filename):
    \"\"\"Procesa el archivo para contar palabras y escribir el resultado.\"\"\"
    start_time = time.time()

    # Leer y procesar el archivo
    content = read_file(input_filename)
    if not content:
        print(f"Error: No se encontraron datos válidos en {input_filename}.")
        return

    word_count = count_words(content)
    elapsed_time = time.time() - start_time

    # Mostrar los resultados
    print(f"\\nResultados para {input_filename}:")
    for word, count in sorted(word_count.items(), key=lambda x: (-x[1], x[0])):
        print(f"{word}: {count}")
    print(f"Grand Total: {sum(word_count.values())}")
    print(f"Tiempo transcurrido: {elapsed_time:.4f} segundos")

    # Escribir resultados en el archivo de salida
    write_results(output_filename, word_count, elapsed_time)

def main():
    \"\"\"Función principal para procesar los archivos especificados.\"\"\"
    files = [
        "TC1.txt", "TC2.txt", "TC3.txt", "TC4.txt", "TC5.txt"
    ]

    base_path = "/content/drive/MyDrive/Pruebas de Software/P3/"
    for file in files:
        input_file = f"{base_path}{file}"
        output_file = f"wordCount_{file.replace('.txt', '')}.txt"
        process_file(input_file, output_file)

if __name__ == "__main__":
    main()
"""

with open('wordCount.py', 'w') as f:
    f.write(program_code)

print("Archivo wordCount.py guardado correctamente.")
```
